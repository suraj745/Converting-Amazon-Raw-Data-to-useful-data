# Cleaning <font color='orange'>Amazon</font> Data and converting it in to useful data


## This data is extracted from <font color='orange'>Amazon</font> using <font color='Red'>Web Scraping</font>

```python
#importing required libraries

import pandas as pd,numpy as np
```

```python
#reading data from csv
df = pd.read_csv('./amazon.csv')
df.head(4)
```

```python
#checking shape of data before cleaning
df.shape
```

```python
#removing unwanted columns 
df=df.iloc[:,[5,6,8,10,12,13,14,15,18]]
```

```python
df.head()
```

```python
#rename column of dataframe
df = df.rename(columns={
    'a-size-base-plus':'device_name',
 'a-icon-alt':'ratings',
   'a-size-base':'reviews',
    'a-offscreen':'price_on_discount',
    'a-color-base':'coupon',
    'a-color-base 2':'date',
    'a-color-base 3':'delivery',
    'a-row':'offer_percentage',
    'a-offscreen 2':'actual_price'
})

```

```python
#checking shape of data
df.shape
```

```python
#checking null values
df.isnull().sum()         # 3 empty rows in date and delivery
                         # and 2 empty rows in actual_price
```

```python
#removing rows from dataframe date column which does not have value
df = df[~(df['date'].isnull())]
```

```python
#checking again null value is remove or not
df.isnull().sum()    #there are two empty rows still in actual price column
```

```python
#checking that actual_price column
df[df['actual_price'].isnull()]  #we can replace that value with price_on_discount bcz offer_perecentage is not there
```

```python
df.head(5) #checking first five rows of data
```

```python
df.head(3) #checking first three row of data
```

```python
# if actual price is empty filling this with price on discount and  
#if there is offer percentage it will add that percent to price on discount and add it to actual price

def fill_actualprice():
    for i in df[df['actual_price'].isnull()].index:
        if df[df.index==i]['offer_percentage'][i] == "Available Services:":
             df.loc[df.index == i,'actual_price']  = int(df[df.index==i]['price_on_discount'][i][1:].replace(',',""))
        else:
             df.loc[df.index == i,'actual_price']  =(int(df['offer_percentage'].strip('()').split(' ')[0].replace('%',"")) * int( df[df.index==i]['price_on_discount'][i][1:].replace(',',""))/100)+int( df[df.index==i]['price_on_discount'][i][1:].replace(',',""))

        
    
              
fill_actualprice()
```

```python
#removing ₹ and ','
df['actual_price'] = df['actual_price'].apply(lambda x: int(x) if str(x).find('₹') else x[1:].replace(',',""))
```

```python
#removing ₹ and ','
df['price_on_discount'] = df['price_on_discount'].apply(lambda x:int(x[1:].replace(',',"")))
```

```python
df.head()
```

```python
#adding category column according to device

def filter_category(x):
    if x.find("Watch") != -1:
        return "watch"
    elif x.find('micro') !=-1:
        return 'sdcard'
    elif x.find('GB') != -1:
        return 'smartphone'
    else:
        return 'earpod'
        
   
df.insert(1,'category',df['device_name'].apply(filter_category))
```
